```python
from bs4 import BeautifulSoup as bs
```

```python
import requests
```

```python
url = 'https://webscraper.io/test-sites/e-commerce/allinone'
```

```python
response = requests.get(url)
```

```python
response
```

```python
data = response.text
```

```python
soup = bs(data,'html.parser')
```

```python
soup.prettify()
```

```python
soup
```

```python
a_tag = soup.find_all('a')
```

```python
a_tag
```

```python
for tag in a_tag:
    print(tag.get('href'))
```
